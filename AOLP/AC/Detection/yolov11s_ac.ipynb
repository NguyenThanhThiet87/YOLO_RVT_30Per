```python
# Chuẩn bị file dữ liệu 
import yaml

# Cấu trúc dữ liệu cho file data.yaml
data_yaml = {
    'train': '/kaggle/input/datasets/nguynthanhthit/aolp-ac/AOLP_AC/AOLP_AC/train/images',
    'val': '/kaggle/input/datasets/nguynthanhthit/aolp-ac/AOLP_AC/AOLP_AC/val/images',
    'nc': 1,
    'names': ['Lisence_Plate']
}

# Ghi file data.yaml vào /kaggle/working/
with open('/kaggle/working/data.yaml', 'w') as f:
    yaml.dump(data_yaml, f)

# Kiểm tra nội dung file
!cat /kaggle/working/data.yaml
```

```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")
model.train(
    data="/kaggle/working/data.yaml",
    epochs=100, imgsz=640,
    # Color
    hsv_h=0.015, hsv_s=0.7, hsv_v=0.4,
    # Hình học
    degrees=0.0, translate=0.10, scale=0.50, shear=0.0, perspective=0.0,
    fliplr=0.5, flipud=0.0,
    # Mix/compose
    mosaic=1.0, mixup=0.1, copy_paste=0.0,  # cutmix dùng mặc định (0 nếu muốn tắt)
    # Khác
    erasing=0.4,
    # Tắt mosaic/mixup ở các epoch cuối cho ổn định
    close_mosaic=10
)
```
